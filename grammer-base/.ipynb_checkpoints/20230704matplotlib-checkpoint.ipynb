```python
import matplotlib
```

```python
x = [1,2,3,4,5]
y = [10,9,8,76,6]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
fig1 = plt.figure(figsize=(5,5))
# 折线图
plt.plot(x, y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig1 = plt.figure(figsize=(5,5))
# 散点图
plt.scatter(x, y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
```

```python
# file_path = './data.xlsx'
# df = pd.read_excel(file_path, sheet_name = "Sheet1")
data = pd.read_excel(io='./data.xlsx')
print(data)
```

```python
x = data.loc[:, 'x']
print(type(x))
print(x)
y = data.loc[:, 'y']
print(type(y))
print(y)
```

```python
# x那一列筛选出对应 y>0 的 x
x = data.loc[:, 'x'][y>0]
print(type(x))
print(x)
```

```python
# 将 excel 中的数据转换为 numpy 中的数据
data_array = np.array(data)
print(type(data_array))
print(data_array)
```

```python
data_new = data + 100
# print(data_new)
data_new.head()
```
